##üóÇÔ∏èImport data: vitalsign

```python
import pandas as pd
vital = pd.read_csv('/content/drive/MyDrive/ckdresearch2023/vitalsign.csv')
vital.head()
```

### üîéUniqueValues 
columns: **OrderItemName and ResultItemName** in vitalsign.csv





```python
import pandas as pd
from collections import Counter

# Extract the values from 'OrderItemName' and 'ResultItemName' columns
order_items = vital['OrderItemName'].values
result_items = vital['ResultItemName'].values

# Count the occurrences of each value in 'OrderItemName' and 'ResultItemName'
order_item_counts = Counter(order_items)
result_item_counts = Counter(result_items)

# Convert the counters to dataframes
order_item_df = pd.DataFrame.from_dict(order_item_counts, orient='index', columns=['Count'])
result_item_df = pd.DataFrame.from_dict(result_item_counts, orient='index', columns=['Count'])

# Sort the dataframes by count in descending order
order_item_df = order_item_df.sort_values(by='Count', ascending=False)
result_item_df = result_item_df.sort_values(by='Count', ascending=False)

# Print the dataframe for 'OrderItemName'
print("Dataframe for 'OrderItemName':")
print(order_item_df)

# Print the dataframe for 'ResultItemName'
print("\nDataframe for 'ResultItemName':")
print(result_item_df)
```

##üóÇÔ∏èImport data: Diagnosis

```python
import pandas as pd
diagnosis = pd.read_csv('/content/drive/MyDrive/ckdresearch2023/Diagnosis.csv')
diagnosis.head()
```

### üîé**UniqueValues**
columns: **ProblemCode, ProblemName, and (ProblemCode,ProblemName)** in Diagnosis.csv 

```python
import pandas as pd
from collections import Counter

# Extract the values from 'ProblemCode' and 'ProblemName' columns
problem_codes = diagnosis['ProblemCode'].values
problem_names = diagnosis['ProblemName'].values

# Count the occurrences of each value in 'ProblemCode', 'ProblemName', and (ProblemCode, ProblemName)
problem_code_counts = Counter(problem_codes)
problem_name_counts = Counter(problem_names)
problem_code_name_counts = Counter(zip(problem_codes, problem_names))

# Create a DataFrame from the counters
diagnosis_df = pd.DataFrame.from_dict(problem_code_counts, orient='index', columns=['ProblemCode_Count'])
diagnosis_df['ProblemName_Count'] = pd.Series(problem_name_counts)
diagnosis_df['ProblemCode, ProblemName_Count'] = pd.Series(problem_code_name_counts)

# Sort the DataFrame by count in descending order
diagnosis_df = diagnosis_df.sort_values(by='ProblemCode_Count', ascending=False)

# Print the merged DataFrame
print("Merged DataFrame:")
print(diagnosis_df)
```

##üóÇÔ∏èImport data: Prescription

```python
import pandas as pd
prescription = pd.read_csv('/content/drive/MyDrive/ckdresearch2023/Prescription.csv')
prescription.head()
```

### **üîéUniqueValues** 
columns: **DrugName, GenericName, and (DrugName,GenericName)** in Prescription.csv

```python
import pandas as pd
from collections import Counter

# Extract the values from 'DrugName' and 'GenericName' columns
drug_names = prescription['DrugName'].values
generic_names = prescription['GenericName'].values

# Combine 'DrugName' and 'GenericName' to create a new column 'DrugName,GenericName'
drug_generic_names = [f"{drug},{generic}" for drug, generic in zip(drug_names, generic_names)]

# Count the occurrences of each value in 'DrugName', 'GenericName', and 'DrugName,GenericName'
drug_name_counts = Counter(drug_names)
generic_name_counts = Counter(generic_names)
drug_generic_name_counts = Counter(drug_generic_names)

# Convert the counters to dataframes
drug_name_df = pd.DataFrame.from_dict(drug_name_counts, orient='index', columns=['DrugName_Count'])
generic_name_df = pd.DataFrame.from_dict(generic_name_counts, orient='index', columns=['GenericName_Count'])
drug_generic_name_df = pd.DataFrame.from_dict(drug_generic_name_counts, orient='index', columns=['DrugName_GenericName_Count'])

# Merge the dataframes into a single dataframe
prescription_merged_df = pd.concat([drug_name_df, generic_name_df, drug_generic_name_df], axis=1)

# Sort the merged dataframe by count in descending order
prescription_merged_df  = prescription_merged_df.sort_values(by='DrugName_Count', ascending=False)

# Print the merged dataframe
print("Merged DataFrame:")
print(prescription_merged_df)
```

```python
prescription_merged_df
```

##üóÇÔ∏èImport data: visit

```python
import pandas as pd
visit = pd.read_csv('/content/drive/MyDrive/ckdresearch2023/Visit.csv')
visit.head()
```

### **üîéUniqueValues** 
columns: **PatientUID, PatientvisitUID, Sex, Age, ENTY, NATNL** in Visit.csv

```python
import pandas as pd
from collections import Counter

# Extract the values from the columns
patient_uids = visit['PatientUID'].values
patient_visit_uids = visit['PatientvisitUID'].values
sexs = visit['Sex'].values
ages = visit['Age'].values
enties = visit['ENTY'].values
natnls = visit['NATNL'].values

# Count the occurrences of each value in the columns
patient_uid_counts = Counter(patient_uids)
patient_visit_uid_counts = Counter(patient_visit_uids)
sex_counts = Counter(sexs)
age_counts = Counter(ages)
enty_counts = Counter(enties)
natnl_counts = Counter(natnls)

# Convert the counters to dataframes
patient_uid_df = pd.DataFrame.from_dict(patient_uid_counts, orient='index', columns=['PatientUID_Count'])
patient_visit_uid_df = pd.DataFrame.from_dict(patient_visit_uid_counts, orient='index', columns=['PatientvisitUID_Count'])
sex_df = pd.DataFrame.from_dict(sex_counts, orient='index', columns=['Sex_Count'])
age_df = pd.DataFrame.from_dict(age_counts, orient='index', columns=['Age_Count'])
enty_df = pd.DataFrame.from_dict(enty_counts, orient='index', columns=['ENTY_Count'])
natnl_df = pd.DataFrame.from_dict(natnl_counts, orient='index', columns=['NATNL_Count'])

# Merge the dataframes into a single dataframe
visit_merged_df = pd.concat([patient_uid_df, patient_visit_uid_df, sex_df, age_df, enty_df, natnl_df], axis=1)

# Print the merged dataframe
print("Merged DataFrame:")
print(visit_merged_df)
```
