###  Dataframe 

```python
data = [15,23,20,30]
col = ['Age']
df =pd.DataFrame(data,columns=col)
df
```

###  Dataframe  list 

```python
data1 = [['','',1200],['','',900],['Iphone 12','',30000]] 
col1 = ['Name','Category','Price']
df1 = pd.DataFrame(data1,columns=col1)
df1
```

###  Dataframe  list  zip
  list  

```python
name = ['','','Iphone 12']
Category	= ['','','']
Price = ['1200','900','30000']
datas = list(zip(name,Category,Price))
cols = ['','','']
df2 = pd.DataFrame(datas,columns=cols)
df2
```

###  Dataframe  dictionary
- key 
- value 

```python
datas1 = [{'Name':'','Category':'','Price':1200},{'Name':'','Category':'','Price':800},{'Name':'Iphone 12','Category':'','Price':30000}]
df3 = pd.DataFrame(datas1)
df3 
```

```python
df3.set_index(['Name'],inplace=True) #  index  Name
```

```python
df3
```

###  Dataframe  series

```python
Name = ['','','Iphone 12']
Category	= ['','','']
Price = ['1200','900','30000']

name_s = pd.Series(Name)
category_s = pd.Series(Category)
price_s = pd.Series(Price)

fram = {'Name':name_s,'Category':category_s,'Price':price_s}
df4 = pd.DataFrame(fram)
df4
```
