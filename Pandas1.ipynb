## Series
Series คือ เป็นโครงสร้างข้อมูลแบบ Array 1 มิติ (คอลัมน์เดียว) ที่เก็บข้อมูลต่างชนิดกันได้ มีส่วนประกอบ 2 ส่วน คือ
- Index ใช้อ้างอิงตำแหน่งของข้อมูล
- Element Value คือ ข้อมูลในแต่ละ index

### การสร้าง Series จาก list และ tuple

```python
data_ls = [10,20,'Kong',15.05,'มะละกอ'] # list
ls = pd.Series(data_ls)
ls
```

```python
data_tp =(10,20,'Kong',15.05,'มะละกอ') # tuple
tp = pd.Series(data_tp)
tp
```

### การสร้าง Series จาก numpy

```python
import numpy as np
```

```python
ndata = np.array(data_ls)
ps = pd.Series(ndata)
ps
```

Array คือ การนำข้อมูลมาอยู่ในกลุ่มเดียวกัน โดยสมาชิกภายใน Array ต้องมี**ชนิดข้อมูลเหมือนกัน**


### การสร้าง Series จาก กำหนดเลข index

```python
item = ['องุ่น','กล้วย','ส้ม'] # list
idx = [50,20,30]
ps1 =  pd.Series(item,index=idx)
ps1
```

### การสร้าง Series จาก Dictionary
data_dict = {key:value,key:value}

```python
data = {'องุ่น':50,'กล้วย':20,'ส้ม':30}
ps2 =  pd.Series(data)
ps2
```

```python
ps2['กล้วย']
```

### การเข้าถึงข้อมูลใน series (แบบช่วงข้อมูล Slice)
ps[star:stop-1]

## Dataframe
Dataframe เป็นโครงสร้างข้อมูลแบบ Array 2 มิติ (ลักษณะเป็นตารางประกอบด้วย 2 คอลัมน์ขึ้นไป) หรือ การนำ Series หลาย ๆ อันมาเรียงต่อกัน เช่น ข้อมูลคะแนนนักเรียน ข้อมูลสินค้า ข้อมูลประชากร เป็นต้น

### การสร้าง Dataframe จาก list และ tuple



```python
data_ls = [10,20,'Kong',15.05,'มะละกอ'] # list
df_ls = pd.DataFrame(data_ls)
df_ls
```

```python
data_tp =(10,20,'Kong',15.05,'มะละกอ') # tuple
df_tp = pd.DataFrame(data_tp)
df_tp
```

### การสร้าง Dataframe แบบกำหนดคอลัมน์

```python
data = [15,23,20,30]
col = ['Age']
df =pd.DataFrame(data,columns=col)
df
```

### การสร้าง Dataframe จาก list หลายมิติ

```python
data1 = [['คีย์บบอร์ด','อุปกรณ์คอม',1200],['ตุ๊กตา','ของเล่น',900],['Iphone 12','มือถือ',30000]] 
col1 = ['Name','Category','Price']
df1 = pd.DataFrame(data1,columns=col1)
df1
```

### การสร้าง Dataframe จาก list ด้วย zip
กาารบีบอัดข้อมูล การนำ list ที่มีลักษณะคล้ายคลีงกันเอามารวมกันในกลุ่มๆหนึ่ง แล้วนำกลุ่มทั้งหมดไปวร้างตาราง

```python
name = ['คีย์บบอร์ด','ตุ๊กตา','Iphone 12']
Category	= ['อุปกรณ์คอม','ของเล่น','มือถือ']
Price = ['1200','900','30000']
datas = list(zip(name,Category,Price))
cols = ['ชื่อสินค้า','หมวดหมู่','ราคา']
df2 = pd.DataFrame(datas,columns=cols)
df2
```

### การสร้าง Dataframe จาก dictionary
- key เป็นคอลัมน์
- value เป็นแถว

```python
datas1 = [{'Name':'คีย์บอร์ด','Category':'อุปกรณ์คอม','Price':1200},{'Name':'ตุ๊กตา','Category':'ของเล่น','Price':800},{'Name':'Iphone 12','Category':'มือถือ','Price':30000}]
df3 = pd.DataFrame(datas1)
df3 
```

```python
df3.set_index(['Name'],inplace=True) # เปลี่ยน index เป็นคอลัมน์ Name
```

```python
df3
```

### การสร้าง Dataframe จาก series

```python
Name = ['คีย์บบอร์ด','ตุ๊กตา','Iphone 12']
Category	= ['อุปกรณ์คอม','ของเล่น','มือถือ']
Price = ['1200','900','30000']

name_s = pd.Series(Name)
category_s = pd.Series(Category)
price_s = pd.Series(Price)

fram = {'Name':name_s,'Category':category_s,'Price':price_s}
df4 = pd.DataFrame(fram)
df4
```

 - แกนที่ 1 แถวหรือ Index ของแต่ละ Dataframe
