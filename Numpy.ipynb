## matrix ค่าคงที่ (Full) & Empty
 - Full อยากจะให้สมาชิกทุกตัวเป็นค่าคงที่อื่นๆ
 - Empty สร้างอาเรย์เปล่าไม่ระบุสมาชิก แต่ระบุจำนวนขนาด (นำขนาดไปใช้)

```python
np.full(5,10) # (จำนวนสมาชิก, ค่าคงที่)
```

```python
np.full((3,3),5) #((row,col),ค่าคงที่)
```

```python
np.full((3,4),2.5) #((row,col),ค่าคงที่)
```

```python
np.empty([2,4])
```

## Flatten & ravel
เปลี่ยนแปลงหลายมิติให้เป็นมิติเดียว
- Flatten ไม่เชื่อมโยงกับข้อมูล เป็นตัวแปรใหม่เลย
- ravel เชื่อมโยงกับข้อมูลก่อนหน้า ถ้าข้อมูลปลี่ยนแปลงข้อมูลเริ่มต้น/ข้อมูลดิบ เปลี่ยนด้วย


```python
x = np.array([[1,2],[3,4],[5,6]])
x
```

```python
x.flatten()
```

```python
x
```

```python
x1 = x.flatten()
x1
```

```python
x1[2]= 8
```

```python
x1
```

```python
x
```

```python
y = x.ravel()
y
```

```python
y[4]=10
```
